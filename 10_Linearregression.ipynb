What is Regression?

Regression is a statistical and machine learning technique used to model and analyze relationships between variables. It predicts a continuous numerical output (dependent variable) based on one or more input features (independent variables). Regression is commonly used in predictive modeling, data analysis, and forecasting.
Key Objectives of Regression

    Prediction: Estimate the value of a target variable based on input features.
    Understanding Relationships: Analyze how changes in input variables impact the output.
    Trend Analysis: Identify patterns or trends in data.

Regression ek statistical technique hai jo kisi dependent variable (output) aur ek ya zyada independent variables (inputs) ke beech ke relationship ko samajhne aur model karne ke liye use hoti hai. Iska main goal hota hai ki input variables ke basis par output variable ko predict kiya ja sake.

Example ke liye:

    Agar aapke paas data ho jisme "ghar ki size" (independent variable) aur "ghar ki price" (dependent variable) ke baare mein information ho, toh regression technique se aap predict kar sakte hain ki ghar ki price kis tarah se size par depend karti hai.

    Isme model ko train karte waqt ek line (ya curve) banayi jaati hai jo data ko sabse acchi tarah fit kare, jise best-fit line kaha jata ha

    Linear Regression ek statistical technique hai jo dependent aur independent variables ke beech ke relationship ko model karta hai. Isme assumption hota hai ki in dono variables ke beech ek linear (straight-line) relationship hota hai.
Definition:

Linear regression ka main goal hota hai ki ek line banayi jaye jo data ke points ke beech ka relationship represent kare, jise best-fit line kaha jata hai. Is line ka use future predictions ke liye kiya jata hai.

Linear Regression Ki Types:

    Simple Linear Regression: Jab sirf ek independent variable (X) aur ek dependent variable (Y) ho.
    Multiple Linear Regression: Jab ek se zyada independent variables (X₁, X₂, ...) ho aur unka effect dependent variable (Y) par analyze karna ho.

    Example:

Maan lijiye, aapke paas data hai jisme "years of experience" (X) aur "salary" (Y) diye gaye hain. Linear regression aapko help karega yeh predict karne mein ki kisi new employee ki salary kitni hogi based on unke experience.
Use Cases:

    Sales Prediction: Sales ko predict karna based on advertising spend.
    Price Prediction: Real estate prices ko predict karna based on property size.
    Trend Analysis: Data trends ko analyze karna over time.

Key Concepts:

    Slope (β₁): Yeh bataata hai ki independent variable (X) ke ek unit increase par dependent variable (Y) mein kitna change hoga.
    Intercept (β₀): Yeh line ka starting point hota hai (jab X = 0).

Linear regression ko simple aur powerful model mana jata hai jab data mein linear relationship ho

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Example data: Calories and Weight
data = {'Calories': [1200, 1500, 1700, 1800, 2000, 2500],
        'Weight': [55, 57, 58, 59, 60, 65]}
df = pd.DataFrame(data)

# Independent variable (X) - Calories consumed
X = df[['Calories']]  # 2D array (for model input)
# Dependent variable (Y) - Body weight
y = df['Weight']

# Create and fit the regression model
model = LinearRegression()
model.fit(X, y)

# Predictions based on the model
y_pred = model.predict(X)

# Print model parameters
print(f"Intercept (β0): {model.intercept_}")
print(f"Slope (β1): {model.coef_[0]}")

# Model evaluation
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared: {r2}")

# Visualize the results
plt.scatter(X, y, color='blue', label='Actual Data')  # Actual data points
plt.plot(X, y_pred, color='red', label='Regression Line')  # Regression line
plt.xlabel('Calories Consumed')
plt.ylabel('Body Weight (kg)')
plt.legend()
plt.title('Simple Linear Regression: Food (Calories) vs Weight')
plt.show()
```

Variance, Residuals, aur Mean Squared Error (MSE) teeno regression analysis aur model evaluation ke interconnected concepts hain. Inka connection samajhna regression ke functioning aur accuracy ko achhe se samajhne mein madad karta hai.
1. Variance aur Residuals ka Connection:

    Variance measure karta hai ki data kitna spread hai. Yeh original data points (dependent variable YY) ke mean (μμ) ke aas-paas ka spread batata hai:
    Variance=1N∑i=1N(yi−μ)2
    Variance=N1​i=1∑N​(yi​−μ)2

    Residuals actual values (yiyi​) aur predicted values (y^iy^​i​) ke beech ka difference hai:
    Residual=yi−y^i
    Residual=yi​−y^​i​

    Key Difference:
        Variance pure data (original values yiyi​) ka mean se deviation measure karta hai.
        Residuals predicted model ke context mein deviation (errors) batate hain.

    Agar residuals ka variance kam hai, iska matlab hai ki model data ko acche se fit kar raha hai.

2. MSE aur Residuals ka Connection:

    Mean Squared Error (MSE) ek evaluation metric hai jo residuals ka squared average leta hai. Formula:
    MSE=1N∑i=1N(yi−y^i)2
    MSE=N1​i=1∑N​(yi​−y^​i​)2

    Yahan:
        yi−y^iyi​−y^​i​ residuals hain.
        MSEMSE residuals ka mean square hai.

    MSE directly residuals ke spread ko represent karta hai. Agar residuals random aur small hain, toh MSE bhi small hoga, jo ek achhe model ka signal hai.

3. Variance aur MSE ka Connection:

    Variance ek baseline metric hai, jo dependent variable (YY) ki variability batata hai.
    Jab hum ek regression model fit karte hain, hum yeh chahte hain ki residuals ka variance original data ke variance se kaafi kam ho. Iska matlab hai ki model ne variance ka kaafi hissa explain kar diya hai.
    Model ka performance assess karne ke liye, MSEMSE aur data ke variance ke beech ka ratio dekhte hain. Isse R2R2 score ya explained variance score calculate hota hai.

Conceptual Summary:

    Variance:
        Data ka natural spread (original data aur mean ke beech).

    Residuals:
        Actual aur predicted values ke beech ka deviation.

    MSE:
        Residuals ka squared average, jo model ke prediction errors ka overall measure hai.

Agar model achha hai, toh:

    Residuals ka spread random aur small hoga.
    MSE chhota hoga, jo batata hai ki residuals ka variance kam hai aur model ne data ka maximum variance explain kiya hai.

# Let's interpret the evaluation metrics:
1. Mean Squared Error (MSE): 0.7747

    Interpretation:
        MSE measures the average of the squared differences between the actual values and the predicted values. It gives us an idea of how much error there is in the predictions.
        A lower MSE indicates better performance, but its absolute value depends on the scale of the data. In this case, 0.7747 is relatively low, suggesting that the model's predictions are somewhat close to the actual values.
        However, we can't judge the model's performance solely by MSE because it doesn't give us a sense of the model's goodness relative to the data. This is where R² and RMSE come in.

2. R-squared (R²): 0.5005

    Interpretation:
        R² represents the proportion of the variance in the dependent variable (y_test) that is predictable from the independent variable (X_test).
        R² = 0.5005 indicates that about 50% of the variance in the dependent variable is explained by the independent variable using this regression model. This suggests that the model is moderately effective but does not explain all the variance.
        An R² close to 1 would indicate a good model fit, while an R² close to 0 suggests a poor fit. Here, 0.5005 is a moderate value, meaning that the model is explaining some but not all of the variability in the target variable. You might want to improve the model or explore more features for better predictive accuracy.

3. Root Mean Squared Error (RMSE): 0.8802

    Interpretation:
        RMSE is the square root of MSE and provides the error in the same units as the dependent variable. It gives a clearer sense of how much the model’s predictions deviate from the actual values.
        RMSE = 0.8802 means that, on average, the predicted values deviate from the actual values by approximately 0.88 units.
        Like MSE, lower RMSE indicates better performance. Since this is in the same units as the dependent variable, it helps you interpret the model’s performance in context. If the RMSE is small relative to the range of values in your target variable, this suggests a good model fit.

R²: Measures the proportion of the variance in the target variable explained by the model; higher values indicate better model fit.
MSE: Represents the average squared difference between actual and predicted values; lower values indicate better model accuracy.
RMSE: The square root of MSE; it shows the average error in the same units as the target variable, with lower values indicating better predictions.
Coefficient: Represents the amount by which the target variable changes for each one-unit change in the independent variable; it shows the strength and direction of the relationship.

R²: Batata hai ki model target variable ki kitni variability ko explain karta hai; zyada value achhe model fit ko dikhati hai.
MSE: Actual aur predicted values ke beech ka average squared difference; kam value model ki accuracy ko dikhati hai.
RMSE: MSE ka square root; yeh target variable ke same units mein average error ko dikhata hai, aur kam value achhi predictions ko dikhati hai.
Coefficient: Batata hai ki target variable har ek unit change par kitna badalta hai independent variable ke liye; yeh relationship ki strength aur direction ko dikhata hai.

```python
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the tips dataset
data = sns.load_dataset('tips')

# Define X (independent variables) and y (dependent variable)
X = data[['total_bill', 'size']]  # Using 'total_bill' and 'size' to predict 'tip'
y = data['tip']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print results
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")
print(f"MSE: {mse}")
print(f"R²: {r2}")
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the tips dataset
data = sns.load_dataset('tips')

# Define X (independent variables) and y (dependent variable)
X = data[['total_bill', 'size']]  # Using 'total_bill' and 'size' to predict 'tip'
y = data['tip']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print results
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")
print(f"MSE: {mse}")
print(f"R²: {r2}")

# Plot Actual vs Predicted Tips
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', label='Predicted vs Actual')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--', label='Ideal Fit')
plt.xlabel('Actual Tips')
plt.ylabel('Predicted Tips')
plt.title('Linear Regression: Actual vs Predicted Tips')
plt.legend()
plt.show()
```
