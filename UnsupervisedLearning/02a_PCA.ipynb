# PCA (Principal Component Analysis) - Definition
Principal Component Analysis (PCA) is a statistical technique used for dimensionality reduction. It transforms a large set of correlated features (variables) into a smaller set of uncorrelated variables called principal components. These components capture the most significant variance in the data, helping to simplify complex datasets while retaining most of the important information.

Key Points:
Goal: To reduce the number of variables (features) while preserving as much variability (information) as possible.
Components: The new variables, called principal components, are linear combinations of the original variables.
Orthogonality: The principal components are orthogonal (uncorrelated) to each other.
Variance: The first principal component captures the largest variance, the second captures the next largest variance, and so on.
Why Use PCA?
Dimensionality Reduction: Reduces the number of features in large datasets, making them easier to visualize, analyze, or use for machine learning models.
Noise Reduction: By keeping only the most important components, PCA can help remove noise from the data.
Improved Model Performance: Reducing features can help prevent overfitting in machine learning models.
Example:
If you have a dataset with many features, like the height, weight, and age of individuals, PCA can combine these features into new ones (principal components) that are easier to work with. The new features might not directly correspond to "height" or "weight" but will capture the underlying structure of the data.

Steps in PCA:
Standardize the data (if needed).
Calculate the covariance matrix to understand how variables are related.
Compute the eigenvalues and eigenvectors of the covariance matrix.
Sort the eigenvectors by their eigenvalues in descending order.
Select the top k eigenvectors (principal components).
Transform the original data to the new basis defined by the selected principal components.
In summary, PCA simplifies complex data by reducing its dimensions, making it easier to analyze, visualize, or build predictive models.

Here are some important key terms related to Principal Component Analysis (PCA):

1. Principal Components:
The new variables formed from linear combinations of the original features. They are uncorrelated and capture the most variance in the data.
Key Point: The first principal component captures the most variance, the second captures the second most, and so on.
2. Eigenvalues:
A measure of the amount of variance that each principal component explains in the data.
Key Point: Larger eigenvalues indicate that a component explains more of the total variance in the data.
3. Eigenvectors:
The directions in the feature space along which the data varies. Eigenvectors define the new axes (principal components) for the transformed data.
Key Point: Each eigenvector corresponds to a principal component, and the coefficients of the eigenvector (loading scores) show how much each feature contributes to the component.
4. Covariance Matrix:
A matrix that represents the covariance (relationships) between all pairs of features in the dataset.
Key Point: PCA begins by calculating the covariance matrix to understand how the features relate to one another.
5. Variance:
A measure of how spread out the values of a feature are. In PCA, variance indicates how much information (or structure) a component captures.
Key Point: PCA aims to capture the maximum variance in the data, reducing dimensions while preserving as much of the original variability as possible.

```python
import seaborn as sns
import pandas as pd

# Load the tips dataset
tips = sns.load_dataset('tips')

# One-hot encoding of categorical features
tips_encoded = pd.get_dummies(tips, drop_first=True)

# Ensure all boolean columns are converted to 1s and 0s
tips_encoded = tips_encoded.astype(int)

# Show the first few rows of the encoded dataset
print(tips_encoded.head())
```

Step 3: Scale the Data
PCA is sensitive to the scale of the data, so we need to standardize the numerical features using StandardScaler to have zero mean and unit variance.

```python
from sklearn.preprocessing import StandardScaler

# Separate the features (X) and target (y)
X = tips_encoded.drop(columns='tip')
y = tips_encoded['tip']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Check the scaled data (first few rows)
print(X_scaled[:5])
```

Step 4: Apply PCA
Now, we will apply PCA to reduce the dimensions. We will reduce the data to 2 principal components to make it easier to visualize.

```python
from sklearn.decomposition import PCA

# Apply PCA with 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame with the PCA components
df_pca = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
df_pca['tip'] = y  # Add the target variable to the PCA DataFrame

# Show the DataFrame with PCA components
print(df_pca.head())
```

Step 5: Visualize the Transformed Data
We'll plot the data before and after PCA. Before PCA, we should visualize the original features, and after PCA, we will visualize the two principal components.

Original Data Scatter Plot (Before PCA):
We will plot two of the original features (total_bill and size) to visualize the data before applying PCA.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of the original data
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='size', data=tips, hue='sex', palette='coolwarm', s=100, alpha=0.7)
plt.title('Original Data (Before PCA)')
plt.xlabel('Total Bill')
plt.ylabel('Size of Party')
plt.show()
```

Scatter Plot After PCA:
Next, we will plot the first two principal components.

```python
# Scatter plot of PCA components
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', data=df_pca, hue='tip', palette='coolwarm', s=100, alpha=0.7)
plt.title('Data After PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Tip')
plt.show()
```

Step 6: Explained Variance
We can check how much of the variance is explained by the first two components to evaluate the effectiveness of PCA.

```python
# Check the explained variance ratio
print("Explained Variance Ratio (PCA1 + PCA2):", pca.explained_variance_ratio_.sum())
```

```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the Tips dataset
tips = sns.load_dataset('tips')

# One-hot encoding of categorical features
tips_encoded = pd.get_dummies(tips, drop_first=True)

# Separate the features (X) and the target (y)
X = tips_encoded.drop('tip', axis=1)  # Features (exclude the target 'tip')
y = tips_encoded['tip']  # Target

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying PCA
pca = PCA(n_components=5)  # Let's try using 5 components
X_pca = pca.fit_transform(X_scaled)

# Explained variance
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()

# Print explained variance and cumulative variance
print(f"Explained variance per component: {explained_variance}")
print(f"Cumulative explained variance: {cumulative_variance}")

# Visualize explained variance
plt.figure(figsize=(8, 5))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.7, label='Explained Variance')
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, color='r', marker='o', label='Cumulative Variance')
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.title('Explained Variance by PCA Components')
plt.legend()
plt.show()

# Visualizing the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.colorbar(label='Tip Amount')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: First Two Principal Components')
plt.show()

# PCA Components (Loadings)
components = pd.DataFrame(pca.components_, columns=X.columns)
print("\nPCA Components (Loadings):")
print(components)
```

Steps Covered:
Data Loading: Loads the tips dataset using Seaborn.
One-Hot Encoding: Converts categorical columns (sex, smoker, day, time) into binary columns using pd.get_dummies().
Scaling: Scales the features using StandardScaler, which is essential before applying PCA to ensure all features have the same scale.
PCA Application: Applies PCA to the scaled data with n_components=5 to extract the first 5 principal components. You can adjust the number of components depending on how much variance you want to retain.
Explained Variance: Prints the explained variance for each principal component and the cumulative variance, showing how much of the data's variability is captured by the selected components.
Variance Plot: Visualizes both the explained variance and cumulative explained variance across the components.
PCA Scatter Plot: Visualizes the first two principal components in a scatter plot, with color representing the tip amount.
PCA Components (Loadings): Prints the loadings, showing how much each original feature contributes to each principal component.
Explanation:
Explained Variance: The output will show the proportion of variance captured by each component. The cumulative variance helps determine how many components to keep based on your desired level of variance retention.
Scatter Plot: The PCA projection helps visualize the data in a reduced-dimensional space (2D in this case) while retaining as much variability as possible.
PCA Components: This tells you the contribution of each original feature (like total_bill, size, sex_male, etc.) to each of the principal components.
