# Data Merge merge hota 2 cheezo ko sma dena 

```python
import pandas as pd

# Creating dataframes
students = pd.DataFrame({
    'Student_ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [14, 15, 14, 15],
    'Grade': [8, 9, 8, 9]
})

scores = pd.DataFrame({
    'Student_ID': [1, 1, 2, 3, 5],
    'Subject': ['Math', 'Science', 'Math', 'Science', 'History'],
    'Marks': [85, 90, 80, 95, 88]
})

teachers = pd.DataFrame({
    'Teacher_ID': ['T1', 'T2', 'T3'],
    'Subject': ['Math', 'Science', 'History'],
    'Teacher_Name': ['Mr. Smith', 'Ms. Brown', 'Mr. Johnson']
})
# merge the data
md =pd.merge(students, scores, on="Student_ID",how="inner")
print(md)
```

# concatination  ikatha karna 
The main difference between merge and concat in pandas lies in how the datasets are combined:

    merge: Combines datasets based on a common column or index (like SQL joins). It aligns rows based on keys.
    concat: Combines datasets along rows or columns without considering keys, stacking them together either vertically (row-wise) or horizontally (column-wise).

```python
# Vertical concatenation Vertical Concatenation (Row-wise)index o or 1 used for rows and columns
concat_data = pd.concat([students, students], ignore_index=True)
print(concat_data)
```

```python
# Horizontal concatenation
concat_data = pd.concat([students, scores], axis=1)
print(concat_data)
```


# Differences Between Merge and Concat in Pandas

| Feature           | Merge                                              | Concat                            |
|-------------------|----------------------------------------------------|-----------------------------------|
| **Purpose**       | Combines data based on keys or common columns.     | Combines data by stacking or side-by-side. |
| **Alignment**     | Aligns rows based on keys.                         | Does not align rows, just appends them.    |
| **SQL Equivalent**| Similar to `JOIN`.                                 | Similar to `UNION`.              |
| **Axes**          | Works row-wise only.                               | Can work row-wise (`axis=0`) or column-wise (`axis=1`). |
| **Custom Keys**   | Requires a common key for alignment.               | Keys are not required.           |
| **Usage**         | `pd.merge(df1, df2, on="key", how="inner")`        | `pd.concat([df1, df2], axis=0)`  |
