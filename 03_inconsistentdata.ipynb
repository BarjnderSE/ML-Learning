```python
import pandas as pd

# Sample data with inconsistencies
data = {
    'Date': ['2020/01/15', '2020-02-20', '2020/03/10', '2020-04-05', '05/05/2020', 
             '2020/06/15', '2020-07-20', '2020-08-12', '09/2020', '2020-10-25', 
             '2020/11/15', '2020-12-10'],
    'Country': ['USA', 'canada', 'UK', 'Germany', 'france', 'USA', 'australia', 'germany', 
                'UK', 'usa', 'Canada', 'france'],
    'Name': ['John Smith', 'Alice Brown', 'Bob Green', 'Carl White', 'Diana Clark', 
             'Eva Black', 'Frank Blue', 'Grace Red', 'Henry Grey', 'Isabella Green', 
             'Jack White', 'Karen Black'],
    'Sales2020': [15000, 12000, 8000, 10000, 13000, 14000, 9000, 10500, 11000, 9500, 
                  12000, 11500],
    'Sales2021': [16000, 12500, 9500, 11000, 13500, 14500, 9500, 10750, 11500, 9800, 
                  12500, 12000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Show the DataFrame
df
```

```python
# 1. Standardize date format
# 2. Convert date to datetime
# Convert the 'Date' column to datetime and standardize the format
# Step 1: Convert 'Date' to datetime and handle different formats
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', dayfirst=False)

# Step 2: Check if any dates became NaT (invalid dates), and handle them if necessary
print(df['Date'].isnull().sum())  # Check how many dates are invalid

# Step 3: Standardize Date format to 'YYYY-MM-DD'
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')

# Step 4: Show the cleaned DataFrame
df.head()
```

```python
import pandas as pd

# Sample data with missing dates
data = {
    'Date': ['2020-01-01', None, '2020-03-01', None, '2020-05-01'],
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Sales2020': [15000, 12000, 8000, 10000, 13000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Ensure the correct date format (you can customize the format string)
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')

# Option 1: Forward fill NaN values (fill with the previous valid date)
df['Date'] = df['Date'].ffill()

# Option 2: Backward fill NaN values (fill with the next valid date)
# df['Date'] = df['Date'].fillna(method='bfill')

# Option 3: Fill NaN values with a specific fallback date (e.g., '2020-01-01')
# df['Date'] = df['Date'].fillna('2020-01-01')

# Show the cleaned DataFrame
print(df)
```

```python
# now fix name of country
# Mapping dictionary for correcting inconsistent country names
country_mapping = {
    'Canad': 'Canada',
    'Germnay': 'Germany',
    'Fraance': 'France',
    'USA': 'United States',
    'UK': 'United Kingdom'
}

# Use the replace function to fix inconsistent country names using the mapping
df['Country'] = df['Country'].replace(country_mapping)
print(df)
```

```python
import pandas as pd

# Sample data with inconsistent names
data = {
    'Date': ['2020-01-01', '2020-01-01', '2020-03-01', '2020-03-01', '2020-05-01'],
    'Country': ['USA', 'Canad', 'UK', 'Germnay', 'Fraance'],
    'Name': ['John Doe', 'Johny Doe', 'Jane Smith', 'Jane Smit', 'Francesca Brown'],
    'Sales2020': [15000, 12000, 8000, 10000, 13000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Clean up column names by stripping extra spaces
df.columns = df.columns.str.strip()

# Check column names
print("Column names after cleanup:", df.columns)

# Ensure the 'Name' column exists
if 'Name' in df.columns:
    # Mapping dictionary for correcting inconsistent names
    name_mapping = {
        'Johny Doe': 'John Doe',
        'Jane Smit': 'Jane Smith',
        'Francesca Brown': 'Francesca Brown'  # No change, just in case
    }

    # Convert the 'Name' column to string type (in case of mixed data types or NaN values)
    df['Name'] = df['Name'].astype(str)

    # Apply mapping to correct names
    df['Name'] = df['Name'].replace(name_mapping)

    # Show the cleaned DataFrame
    print(df)
else:
    print("Column 'Name' does not exist in the DataFrame!")
```

```python
# drop duplicates
df = df.drop_duplicates()
df.head(10)
```

```python
import numpy as np
import pandas as pd

# Sample data
data = {
    'Date': ['2020-01-01', '2020-02-01', '2020-03-01', '2020-04-01', '2020-05-01'],
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Name': ['John Doe', 'Jane Smith', 'John Doe', 'Jane Smith', 'Francesca Brown'],
    'Sales2020': [15000, 12000, 8000, 10000, 13000],
    'Sales2021': [16000, 11000, 8500, 12000, 12500]
}

# Create DataFrame
df = pd.DataFrame(data)

# Show original DataFrame
print("Original DataFrame:")
print(df)

# Set Sales2021 to NaN where Sales2021 is less than or equal to Sales2020
df.loc[df['Sales2021'] <= df['Sales2020'], 'Sales2021'] = np.nan

# Show updated DataFrame with NaNs
print("\nUpdated DataFrame with NaNs where difference is not positive:")
print(df)
```
