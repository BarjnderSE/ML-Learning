```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Actual and predicted labels
y_true = ['Spam', 'Not Spam', 'Not Spam', 'Spam', 'Spam']
y_pred = ['Spam', 'Spam', 'Not Spam', 'Not Spam', 'Spam']

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred, labels=['Spam', 'Not Spam'])
print("Confusion Matrix:\n", cm)

# Calculate Metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, pos_label='Spam')
recall = recall_score(y_true, y_pred, pos_label='Spam')
f1 = f1_score(y_true, y_pred, pos_label='Spam')

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("spam.csv", encoding="latin-1")

# Rename columns
data.columns = ["label", "message"]
data['label'] = data['label'].map({"ham": 0, "spam": 1})

# Prepare data
X = data["message"]
y = data["label"]

# Vectorize text data
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Random Forest Classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Display the Confusion Matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Spam", "Spam"])
disp.plot(cmap=plt.cm.Blues)  # Optional: You can change the color map if you like
plt.title("Confusion Matrix")
plt.show()

# Additional performance metrics
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

To show the Confusion Matrix with formulas, let's first review the key components of the matrix and then display it using a formula.
Confusion Matrix Components:

The confusion matrix for a binary classification problem (like spam vs. not spam) looks like this:
[True Negative (TN)False Positive (FP)False Negative (FN)True Positive (TP)]
[True Negative (TN)False Negative (FN)​False Positive (FP)True Positive (TP)​]

Where:

    True Negative (TN): Correctly predicted as not spam.
    False Positive (FP): Incorrectly predicted as spam when it is not spam.
    False Negative (FN): Incorrectly predicted as not spam when it is spam.
    True Positive (TP): Correctly predicted as spam.

Common Metrics from Confusion Matrix:

From the confusion matrix, we can calculate the following metrics:

    Accuracy: The proportion of correct predictions (both True Positives and True Negatives) to the total number of predictions.
    Accuracy=TP+TNTP+TN+FP+FN
    Accuracy=TP+TN+FP+FNTP+TN​

    Precision (also called Positive Predictive Value): The proportion of correctly predicted spam messages to all predicted as spam.
    Precision=TPTP+FP
    Precision=TP+FPTP​

    Recall (also called Sensitivity or True Positive Rate): The proportion of correctly predicted spam messages to all actual spam messages.
    Recall=TPTP+FN
    Recall=TP+FNTP​

    F1-Score: The harmonic mean of Precision and Recall.
    F1=2×Precision×RecallPrecision+Recall
    F1=2×Precision+RecallPrecision×Recall​

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
data = pd.read_csv("spam.csv", encoding="latin-1")

# Rename columns
data.columns = ["label", "message"]
data['label'] = data['label'].map({"ham": 0, "spam": 1})

# Prepare data
X = data["message"]
y = data["label"]

# Vectorize text data
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Random Forest Classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Calculate metrics
TP = cm[1, 1]
TN = cm[0, 0]
FP = cm[0, 1]
FN = cm[1, 0]

# Accuracy, Precision, Recall, F1 Score
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * (precision * recall) / (precision + recall)

print(f"\nAccuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1_score:.4f}")

# Display the Confusion Matrix with formulas
plt.figure(figsize=(8, 6))

# Plot Confusion Matrix
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Not Spam", "Spam"], yticklabels=["Not Spam", "Spam"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Show formulas
print("\nFormulas:")
print("1. Accuracy = (TP + TN) / (TP + TN + FP + FN)")
print(f"   Accuracy = ({TP} + {TN}) / ({TP} + {TN} + {FP} + {FN}) = {accuracy:.4f}")
print("2. Precision = TP / (TP + FP)")
print(f"   Precision = {TP} / ({TP} + {FP}) = {precision:.4f}")
print("3. Recall = TP / (TP + FN)")
print(f"   Recall = {TP} / ({TP} + {FN}) = {recall:.4f}")
print("4. F1-Score = 2 * (Precision * Recall) / (Precision + Recall)")
print(f"   F1-Score = 2 * ({precision:.4f} * {recall:.4f}) / ({precision:.4f} + {recall:.4f}) = {f1_score:.4f}")
```

Explanation of the Code:

    Confusion Matrix Calculation: The confusion matrix is computed using confusion_matrix() function, which returns a 2x2 matrix with the counts of TP, TN, FP, and FN.

    Metrics Calculation:
        Accuracy, Precision, Recall, and F1-score are calculated based on the values from the confusion matrix.
        These metrics are printed out with the actual calculations based on the matrix values.

    Confusion Matrix Plot:
        The confusion matrix is displayed as a heatmap using seaborn's heatmap function.
        annot=True ensures that the matrix cells are annotated with the values.
        We use cmap="Blues" to set the color scheme for the matrix display.

    Formula Display:
        The formulas for Accuracy, Precision, Recall, and F1-Score are printed along with the actual calculations for the given values of TP, TN, FP, and FN.

Output:

    Confusion Matrix as a heatmap.
    Formulas printed in the console with calculations based on the confusion matrix values.
    Performance Metrics like accuracy, precision, recall, and F1-score.

Example Confusion Matrix:

For example, if the confusion matrix looks like this:
[50 2 ]
[5 43​]

    True Negative (TN) = 50
    False Positive (FP) = 2
    False Negative (FN) = 5
    True Positive (TP) = 43

The formulas will give the following results:

    Accuracy = (50 + 43) / (50 + 2 + 5 + 43) = 93/100 = 0.93
    Precision = 43 / (43 + 2) = 0.955
    Recall = 43 / (43 + 5) = 0.896
    F1-Score = 2 * (0.955 * 0.896) / (0.955 + 0.896) = 0.924
