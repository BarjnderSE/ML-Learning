# Discretization in Data Preprocessing

Discretization is the process of converting continuous data into discrete intervals (or bins). This is especially useful when you want to:

    Simplify numerical data for analysis.
    Convert continuous features into categorical features for certain machine learning algorithms.

1. Methods of Discretization

    Equal Width Binning: Divides the data range into equal-width intervals.
    Equal Frequency Binning: Divides the data such that each bin has an equal number of samples.
    Custom Binning: You define the bin edges manually based on domain knowledge.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset('titanic')

# Drop NaN values in 'age' for simplicity
df = df[['age']].dropna()

# Original age distribution
sns.histplot(df['age'], bins=30, kde=True)
plt.title("Original Age Distribution")
plt.show()
```

# 1. Equal Width Binning

```python
# Equal width binning into 4 bins
df['Age_Equal_Width'] = pd.cut(df['age'], bins=4, labels=['Young', 'Adult', 'Middle-Aged', 'Senior'])

print(df[['age', 'Age_Equal_Width']].head())

# Visualize
df['Age_Equal_Width'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Equal Width Binning")
plt.xlabel("Age Groups")
plt.ylabel("Count")
plt.show()
```

# 2. Equal Frequency Binning

```python
# Equal frequency binning into 4 bins
df['Age_Equal_Freq'] = pd.qcut(df['age'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

print(df[['age', 'Age_Equal_Freq']].head())

# Visualize
df['Age_Equal_Freq'].value_counts().plot(kind='bar', color='orange')
plt.title("Equal Frequency Binning")
plt.xlabel("Age Quartiles")
plt.ylabel("Count")
plt.show()
```

# 3. Custom Binning

```python
# Custom bins
bins = [0, 18, 35, 60, 80]
labels = ['Child', 'Young Adult', 'Adult', 'Senior']
df['Age_Custom'] = pd.cut(df['age'], bins=bins, labels=labels)

print(df[['age', 'Age_Custom']].head())

# Visualize
df['Age_Custom'].value_counts().plot(kind='bar', color='green')
plt.title("Custom Binning")
plt.xlabel("Custom Age Groups")
plt.ylabel("Count")
plt.show()
```



Discretization helps in simplifying the interpretation of data, making it more accessible for certain machine learning models and analyses.
